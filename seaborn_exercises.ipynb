```python
import pandas as pd
from pydataset import data
import seaborn as sns
```

```python
# Use the iris database to answer the following quesitons:

# 1) What does the distribution of petal lengths look like?

petals = data('iris')
print(petals.head())



sns.histplot(petals['Petal.Length'])
```

```python
# Is there a correlation between petal length and petal width?

petals['Petal.Length'].corr(petals['Petal.Width'])
sns.relplot(x = petals['Petal.Length'], y = petals['Petal.Width'])

# Yes, high positive correlation
```
