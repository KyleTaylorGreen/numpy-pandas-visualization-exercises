```python
import matplotlib.pyplot as plt
import math

plt.figure(figsize=(8,8))

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

xvalue = []
x = 0.00000000001
while x < 1000:
    xvalue.append(x*.1)
    x += 1
ylist = [
        [0 * n + 1 for n in xvalue], 'O(1)',
        [math.log(n) for n in xvalue], "O(log n)",
        [n for n in xvalue],  "O(n)",
        [n * math.log(n) for n in xvalue], "O(n log n)",
        [n ** 2 for n in xvalue], "O(n^2)",
        [2 ** n for n in xvalue], "O(2^n)",
        [math.factorial(int(round(n))) for n in range(15)], "O(n!)",
        [n ** n for n in xvalue], "O(n^n)"
]

plt.ylim(-5, 20)
plt.xlim(0, 10)
for x in range(1, len(ylist), 2):
    if ylist[x] == 'O(n!)':
        plt.plot([x for x in range(15)], ylist[x-1], label = ylist[x])
    else:
        plt.plot(xvalue, ylist[x-1], label = ylist[x])
    
    
plt.legend()
plt.show()
```
