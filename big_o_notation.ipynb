```python
import matplotlib.pyplot as plt
import math

plt.figure(figsize=(6,6))

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

xvalue = [x for x in range(1, 100)]
ylist = [
        [0 * n + 1 for n in xvalue], 'O(1)',
        [math.log(n) for n in xvalue], "O(log n)",
        [n for n in xvalue],  "O(n)",
        [n * math.log(n) for n in xvalue], "O(n log n)",
        [n ** 2 for n in xvalue], "O(n^2)",
        [2 ** n for n in xvalue], "O(2^n)",
        [math.factorial(n) for n in xvalue], "O(n!)",
        [n ** n for n in xvalue], "O(n^n)"
]

plt.ylim(-5, 100)
plt.xlim(0, 50)
for x in range(1, len(ylist), 2):
    plt.plot(xvalue, ylist[x-1], label = ylist[x])
    
plt.legend()
plt.show()
```
