```python
# 1) y = x**2 - x + 2

import matplotlib.pyplot as plt
from importlib import reload

xvalue = [x for x in range(10)]
yvalue = [x**2 - x + 2 for x in xvalue]

plt.plot(xvalue, yvalue)
plt.annotate('(0, 0)', xy= (0,0), xytext=(0,5))
plt.show()


```

```python
# 2a) y = sqrt(x)

xvalue = [x for x in range(100)]
yvalue = [x**(1/2) for x in xvalue]
plt = reload(plt)

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')

plt.show()
```

```python
# 2b) y = x**3


xvalue = [x for x in range(5)]
yvalue = [x**3 for x in xvalue]
plt = reload(plt)

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.ylabel('$x^{3}$')

plt.show()
```

```python
# 2c) y = 2**x

xvalue = [x for x in range(5)]
yvalue = [2**x for x in xvalue]
plt = reload(plt)

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.ylabel('$2^{x}$')

plt.show()
```

```python
# 2d) y = 1/(x + 1)

xvalue = [x for x in range(10)]
yvalue = [(1/(x+1)) for x in xvalue]
plt = reload(plt)

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.ylabel(r'$\dfrac{1}{x+1}$')
```

```python
# 3) combine previous four figures into one figure w 4 subplots

plt.figure(figsize=(12,12))

# first subplot
plt.subplot(221)
xvalue = [x for x in range(100)]
yvalue = [x**(1/2) for x in xvalue]

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.title('$\sqrt{x}$')

# second subplot
plt.subplot(222)
xvalue = [x for x in range(5)]
yvalue = [x**3 for x in xvalue]

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.title('$x^{3}$')

# third subplot
plt.subplot(223)
xvalue = [x for x in range(5)]
yvalue = [2**x for x in xvalue]

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.title('$2^{x}$')

# 4th subplot
plt.subplot(224)
xvalue = [x for x in range(10)]
yvalue = [(1/(x+1)) for x in xvalue]

plt.plot(xvalue, yvalue)
plt.xlabel('$x$')
plt.title('$\dfrac{1}{x+1}$')
```

```python
# 4) Combine the figures you created in the last step into one figure 
#    where each of the 4 equations has a different color for the points. 
#    Be sure to include a legend and an appropriate title for the 
#.   figure.
plt = reload(plt)

xvalue = range(4)
y1 = [x**(1/2) for x in xvalue]
y2 = [x**3 for x in xvalue]
y3 = [2**x for x in xvalue]
y4 = [(1/(x+1)) for x in xvalue]

plt.plot(xvalue, y1, c= 'blue', label = '$\sqrt{x}$')
plt.plot(xvalue, y2, c= 'red', label = '$x^{3}$')
plt.plot(xvalue, y3, c= 'orange', label = '$2^{x}$')
plt.plot(xvalue, y4, c= 'purple', label = r'$\frac{1}{x+1}$')
plt.title('All Graphs')
plt.legend()


plt.show()
```
